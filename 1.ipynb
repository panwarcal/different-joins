```python
import pandas as pd
import numpy as np
```

```python
records_1 = pd.read_excel("Records Matching Task.xlsx", sheet_name=0)
print(records_1.shape)
records_1.head()
```

```python
records_2 = pd.read_excel("Records Matching Task.xlsx", sheet_name=1)
print(records_2.shape)
records_2.head()
```

```python
records = pd.merge(records_1, records_2, on=('Order ID', 'Product ID'), suffixes=('_1', '_2'))
print(records.shape)
records.head(100)
```

```python
records_left = pd.merge(records_1, records_2, how='left', on=['Order ID', 'Product ID'], suffixes=('_1', '_2'), indicator=True)
records_left[records_left['_merge'] == 'left_only']
```

```python
records_right = pd.merge(records_1, records_2, how='right', on=('Order ID', 'Product ID'), suffixes=('_1', '_2'), indicator=True)
records_right[records_right['_merge'] == 'right_only']
```

```python
records_outer = pd.merge(records_1, records_2, how='outer', on=('Order ID', 'Product ID'), suffixes=('_1', '_2'), indicator=True)
# records_right[records_right['_merge'] == 'right_only']
records_outer
```
